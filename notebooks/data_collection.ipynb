```python
import os
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
DRIVE_PATH = "/content/drive/My Drive/BARKAI/raw_data"
os.makedirs(DRIVE_PATH, exist_ok=True)
```

```python
while True:
        file_name = input("Enter the filename")
        if not file_name:
            print("Filename cannot be empty. Please try again.")
            continue

        text_content = input("Paste text")
        full_save_path = os.path.join(DRIVE_PATH, file_name)
        with open(full_save_path, "w", encoding="utf-8") as f:
          f.write(text_content)
          print(f"Successfully saved '{file_name}' to Google Drive")

        another_file = input("add another file? (y/n): ").lower().strip()
        if another_file != 'y':
            break

print(os.listdir(DRIVE_PATH))
```
